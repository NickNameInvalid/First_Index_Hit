```python
# load the term lexicon to ignore the query that contains term not occur in CW09B
term_lexicon_file = open("CW09B.fwd.terms", "r", encoding='utf-8')
term_dic = {}
line = term_lexicon_file.readline()
while line:
    line = line.replace("\n", "")
    term_dic[line] = True
    line = term_lexicon_file.readline()
term_lexicon_file.close()
```

```python
# we need to filter those test queries containing terms not occuring in collection
aol_testing_source_file_path = "..\\Testing Part\\aol_testing.txt"
aol_testing_source_file = open(aol_testing_source_file_path, "r", encoding='utf-8')

#aol_testing_target_file_path = "query log for PISA\\aol_testing_filtered.txt"
aol_testing_target_file_path = "query log for PISA\\aol_testing.txt"
aol_testing_target_file = open(aol_testing_target_file_path, "w",encoding='utf-8')

long_query_num = 0
invalid_query_num = 0

query_line = aol_testing_source_file.readline()
while query_line:
    query_line = query_line.replace("\n", "")
    colon_pos = query_line.index(":")
    query_index = query_line[:colon_pos]
    query_content = sorted(list(set(query_line[colon_pos + 1:].split())))
    length = len(query_content)
    
    # too long, skip this query
    if length > 12:
        long_query_num += 1
        query_line = aol_testing_source_file.readline()
        continue
    
    valid = True
    # if any one term not occur, ignore that query
    if "20kine" in query_content:
        print("Yes!")
        print(query_content)
        temp_term = stemmer.stem("20kine")
        print("Temp Term:", temp_term)
        print("Wierd:", temp_term in term_dic)
    for term in query_content:
        term = stemmer.stem(term)
        if term not in term_dic:
            valid = False
            invalid_term = term
            break
    if valid == True:
        aol_testing_target_file.write(query_index + ":" + " ".join(query_content) + "\n")
    else:
        print("Invalid query:", query_index + ":" + " ".join(query_content), "Term: " + invalid_term)
        invalid_query_num += 1
    
    query_line = aol_testing_source_file.readline()
    
aol_testing_source_file.close()
aol_testing_target_file.close()
print("Number of Long Queries:", long_query_num)
print("Number of Invalid Queries:", invalid_query_num)
```

```python
def add_index_for_query_file(file_name):
    directory_name = "query log for PISA"
    
    cut_off_idx = file_name.index("_")
    output_file_name = file_name[:cut_off_idx] + "_queries.txt"
    
    file_reader = open(file_name, "r", encoding='utf-8')
    output_file_writer = open(directory_name + "/" + output_file_name, "w", encoding='utf-8')
    
    line_no = 1
    line = file_reader.readline()
    while line:
        line_list = line.split()
        valid = True
        # if any one term not occur, ignore that query
        for term in line_list:
            term = stemmer.stem(term)
            if term not in term_dic:
                valid = False
                break
        if valid == True:
            output_file_writer.write(str(line_no) + ":" + line)
            line_no += 1
        else:
            print("Invalid query:", str(line_no) + ":" + line)
        line = file_reader.readline()
        
    file_reader.close()
    output_file_writer.close()
    print("Processing Ends.")
    print("*******************************************")
```

```python
start_time = datetime.now()
add_index_for_query_file("single_intersection.txt")
add_index_for_query_file("duplet_intersection.txt")
add_index_for_query_file("triplet_intersection.txt")
add_index_for_query_file("quadruplet_intersection.txt")
add_index_for_query_time = datetime.now() - start_time
print("Time to Add Index for Queries:", str(add_index_for_query_time.total_seconds()) + "s")
```
